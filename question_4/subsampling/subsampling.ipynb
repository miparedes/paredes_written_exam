```python
import pandas as pd
import numpy as np
import random
```

```python
data_df = pd.read_csv("example-dataset.tsv", sep ="\t" )
data_df
```

```python
data_df["strain_name"] =data_df.strain.astype(str)+ "_" + data_df.region.astype(str) + "_" + data_df.date.astype(str)
```

```python
data_df.region.value_counts()
```

```python
unique_locations = data_df.region.unique().tolist(); unique_locations
```

```python
def weighted_subsampling(weights_list):    
    temp_seqs = data_df.copy()
    temp_cases = pd.DataFrame()
    final_df = pd.DataFrame(columns = ["strain", "region", "date", "sequence"])
    sampled_locations = random.choices(population = unique_locations, weights = weights_list, k = 1000)
    #print(sampled_locations)
    for locations in sampled_locations:
        try:
            subset_seqs = pd.DataFrame()
            subset_seqs = temp_seqs[temp_seqs.region == locations].sample(1)
            temp_seqs.drop(subset_seqs.index, inplace = True)
            #print(subset_seqs)
            final_df = final_df.append(subset_seqs, ignore_index=True)
        except ValueError:
            continue 
    return final_df
```

```python
weights_list = [0.5, 0.5]
equal_df = weighted_subsampling(weights_list)
```

```python
equal_df.region.value_counts()
```

```python
weights_list = [0.9, 0.1]
biased_df = weighted_subsampling(weights_list)

biased_df
```

```python
weights_list = [0.9, 0.1]
biased_df = biased_df(weights_list)

biased_df
```

```python
biased_df.region.value_counts()
```

```python
biased_df = biased_df[["strain_name", "sequence"]]
equal_df = equal_df[["strain_name", "sequence"]]
```

```python
biased_df.to_csv("biased_samples.txt", sep = "\t", index = False, header = False)
equal_df.to_csv("equal_samples.txt", sep = "\t", index = False, header = False)
```
